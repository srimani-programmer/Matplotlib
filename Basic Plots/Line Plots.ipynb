## Constructing the Basic Line Plots with Italy Earthquake DataSets.

```python
# Importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('1901-2000.csv')
```

```python
abu_dataset = dataset.head(11)
```

## Drawing a line Plot between Max.Mean Temp and Mean Rain Fall

```python
plt.plot(abu_dataset['Mean Temperature in degree C - Maximum'],abu_dataset['Mean Rainfall in mm'])
plt.xlabel('Mean Temperature in degree C - Maximum')
plt.ylabel("Mean Rainfall in mm")
plt.show()
```

## Comparison between two different Areas

```python
dataset.tail(50)
```

```python
vijayawada_data = dataset[1308:1320]
vishakapatnam_data = dataset[1320:1332]
```

```python
plt.plot(vijayawada_data['Mean Temperature in degree C - Maximum'],vijayawada_data['Mean Rainfall in mm'], label="Vijayawada Analysis")
plt.plot(vishakapatnam_data['Mean Temperature in degree C - Maximum'],vishakapatnam_data['Mean Rainfall in mm'],label="vishakapatnam Analysis")
plt.xlabel('Mean Temperature in degree C - Maximum')
plt.ylabel('Mean Rainfall in mm')
plt.legend()
plt.show()
```

## Introduction to Scatter Plots.

```python
# Importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('1901-2000.csv')
```

```python
dataset.head(50)
```

```python
# Extarcting the Required data
agra_data = dataset[24:36]
ahmedabad_data = dataset[36:48]
```

### Drawing the Scatter Plot Between Max.Temperature and Mean Rainfall.

```python
plt.scatter(agra_data['Mean Temperature in degree C - Maximum'],agra_data['Mean Rainfall in mm'],label="Agra Analysis")
plt.scatter(ahmedabad_data['Mean Temperature in degree C - Maximum'],agra_data['Mean Rainfall in mm'],label='Ahmedabad Analysis')
plt.xlabel('Mean Temperature in degree C - Maximum')
plt.ylabel('Mean Rainfall in mm')
plt.legend(bbox_to_anchor=(0.01, 0.8), loc=6, borderaxespad=0.1) # To change the position of the Legend.
plt.show()
```

### Scatter Plot between  Max & Min with Rainfall in agra. 

```python
plt.scatter(agra_data['Mean Temperature in degree C - Maximum'],agra_data['Mean Rainfall in mm'],marker='*',label="Max and Mean Rainfall")
plt.scatter(agra_data['Mean Temperature  in degree C - Minimum'],agra_data['Mean Rainfall in mm'],marker='o',label="Min and Rainfall")
plt.title('Agra Rainfall Analysis')
plt.xlabel('Temperature in degree C')
plt.ylabel('Mean Rainfall in mm')
plt.legend()
plt.savefig('Agra_Analysis.png')
plt.show()
```
